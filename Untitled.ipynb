```python
# Download project data
import pandas as pd
```

```python
# Data cleaning procedure shown here: https://github.com/NeilCollinsMS/CTU-13-Classification/blob/main/CTU_13_Classification_(Aggregate).ipynb

ctu = pd.read_csv('balanced_df.csv')
```

```python
ctu
```

```python
small_ctu = ctu.sample(frac = 0.005, replace = False)
```

```python
# fit a pipeline
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
y = small_ctu.Malicious
features = ['Dur', 'TotPkts', 'TotBytes', 'SrcBytes', 'arp', 'esp', 'gre', 'icmp', 'igmp', 'ipnip', 'ipv6', 'ipv6-icmp', 'ipx/spx', 'llc', 'pim', 'rarp', 'rsvp', 'rtcp', 'rtp', 'tcp', 'udp', 'udt', 'unas']
X = small_ctu[features].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
pipe = Pipeline([('impute', SimpleImputer()), ('scaler', StandardScaler()), ('svc', SVC())])
pipe.fit(X_train, y_train)
```

```python
import pickle
with open('pipe.pickle', 'wb') as f:
    pickle.dump(pipe, f, pickle.HIGHEST_PROTOCOL)
```
